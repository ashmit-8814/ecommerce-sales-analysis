```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.groupby("Region")["Revenue"].sum().plot(kind="bar")
plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby("Product")["Quantity"].sum().sort_values().plot(kind="barh")
plt.title("Top Selling Products")
plt.xlabel("Quantity Sold")
plt.show()
```

```python
sns.barplot(x="Category", y="Revenue", data=df)
plt.title("Revenue by Category")
plt.show()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Month"] = df["Order_Date"].dt.month
```

```python
df.groupby("Month")["Revenue"].sum().plot(kind="line", marker="o")
plt.title("Monthly Revenue Trend")
plt.ylabel("Revenue")
plt.show()
```

# 📊 E-commerce Sales Analysis - Insights

## 🔹 Regional Performance
- West and North regions generate the highest revenue.
- East region contributes the least.

## 🔹 Product Performance
- Laptops and mobiles are the top-selling products.
- Accessories like mouse and keyboard have lower sales.

## 🔹 Category Analysis
- Electronics category dominates overall revenue.
- Furniture contributes less but has high-value items.

## 🔹 Sales Trend
- Sales show an increasing trend over months.
- Possible reason: seasonal demand or growing customer base.

## 🔹 Business Recommendation
- Focus more on Electronics inventory.
- Increase marketing in low-performing regions.
- Promote high-selling products like laptops and mobiles.
